```python
dataset2['frequency'] = dataset2.tags.apply(lambda n: len(n.split('>')))
dataset2['frequency']= dataset2['frequency']-1
```

```python
#1. Determine the Number of Tags Per Question
dataset2[['id','tags','frequency']]
```

```python
countfrequency = {}

for item in dataset2['frequency']:
   if item in countfrequency:
      countfrequency[item] += 1
   else:
      countfrequency[item] = 1

totaltags = list(countfrequency.keys())
freevalues = list(countfrequency.values())
plt.bar(totaltags, freevalues, color ='black') 
plt.xlabel("number of tags")
plt.ylabel("Freq")
plt.show()
```
